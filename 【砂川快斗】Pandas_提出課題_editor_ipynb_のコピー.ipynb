### 1.2 「次のlist_1をデータフレームに変換し、列名をそれぞれ['名前', '職業', '年齢', '身長', '体重']としてください。また、そのデータフレームをdf_1という名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「Series型, DataFrame型の作成方法」節
</details>

```python
# こちらのセルを実行してください。
list_1 = [['たかし', '学生', 17, 170, 60], 
          ['かな', '学生', 15, 150, 43],
          ['はると', '社会人', 25, 172, 65],
          ['もも', '社長', 30, 158, 48],
          ['まい', '社会人', 23, 163, 51],
          ['おさむ', '専門学生', 19, 175, 68],
          ['あきら', '大学生', 20, 169, 63]]

```

### 1.3 「1.2と同様に、次のlist_2をデータフレームに変換し、列名をそれぞれ['名前', '職業', '年齢', '身長', '体重']としてください。また、そのデータフレームをdf_2という名前の変数に代入して表示してください。」

</details>

```python
# こちらのセルを実行してください。
list_2 = [['そら', np.nan, 20, 173, 61], 
          ['あやか', '学生', 19, 165, np.nan]]
```

```python
df_2 = pd.DataFrame(list_2, columns=['名前', '職業', '年齢', '身長', '体重'])
df_2
```

### 1.12 「次のlist_3をデータフレームに変換し、列名をそれぞれ['名前', '性別']としてください。また、そのデータフレームをdf_3という名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「Series型, DataFrame型の作成方法」節
</details>

```python
# こちらのセルを実行してください。
list_3 = [['たかし', '男性'], ['かな', '女性'], ['はると', '男性'], ['もも', '女性'], ['まい', '女性'], ['そら', '男性'], ['あきら', '男性']]
```

```python
df_3 = pd.DataFrame(list_3, columns=['名前', '性別'])
df_3
```

### 1.13 「df_concatenatedとdf_3を「名前」列で横に結合してください。結合の際には、いずれかのデータフレームに含まれている行を全て使って、結合してください。その横に結合したデータフレームをdf_mergedという名前の変数に代入して、表示してください。

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「テーブル結合方法」節
</details>

```python
df_merged = pd.merge(df_3, df_concatenated, on='名前', how='outer')
df_merged
```

### 1.14 「df_mergedの中で、欠損値が含まれている行を全て削除し、df_merge_dropedという名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「加工処理」節
</details>

```python
df_merge_droped = df_merged.dropna()
df_merge_droped
```

### 1.15 「df_merged_dropedに、「性別」列が男性なら0・女性なら1のように対応させた「性別_int」列を新たに追加してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「テキストデータの加工」節
</details>

```python
df_merge_droped['性別_int'] = df_merge_droped['性別'].map({'男性': 0, '女性': 1})
df_merge_droped
```

### 1.16 「df_merged_dropedの「職業」列に「学生」という文字が含まれている行を全て抽出して、df_studentsという名前の変数に代入して表示してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Pandasの「テキストデータの加工」節
</details>

```python
df_students = df_merge_droped[df_merge_droped['職業'].apply(lambda x: '学生' in x)]
df_students
```
