```python
import japanize_matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1.3 「次のコードを実行してください。」

```python
# こちらのセルを実行してください。
data = pd.DataFrame()
np.random.seed(100)
data['身長'] = np.random.normal(170, 7, 100)
data['クラス'] = ['a', 'b', 'c', 'd'] * 25
data['成績'] = [2, 5, 4, 1, 4, 3, 4, 2, 4, 5] * 10
data['体重'] = np.random.normal(60, 3, 100)

data
```

### 1.4 「dataの『体重』列の分布を、箱ひげ図によって可視化してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」節
</details>

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(data['体重'])

plt.show()
```

### 1.5 「クラス別の『体重』列の平均値を、棒グラフによって可視化してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」節
</details>

```python
fig = plt.figure()
ax = fig.add_subplot(111)

class_weight_median = data.groupby('クラス').median()

ax.bar('a', class_weight_median.loc['a', '体重'])
ax.bar('b', class_weight_median.loc['b', '体重'])
ax.bar('c', class_weight_median.loc['c', '体重'])
ax.bar('d', class_weight_median.loc['d', '体重'])

plt.show()
```

### 1.6 「身長と体重の関係を、散布図によって可視化してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」節
</details>

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(data['身長'], data['体重'])

plt.show()
```

### 1.7 「1.6の散布図と同様のものを描画してください。その際に、figsizeを(12, 6)のように指定してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
ax.scatter(data['身長'], data['体重'])

plt.show()
```

### 1.8 「1.7の散布図と同様のものを描画してください。その際に、タイトルを『身長と体重の関係』、x軸のラベルを『身長』、y軸のラベルを『体重』のように指定してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」・「各種グラフのオプション設定方法」節
</details>

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, xlabel='身長', ylabel='体重', title='身長と体重の関係')
ax.scatter(data['身長'], data['体重'])

plt.show()
```

### 1.9 「x軸にインデックス、y軸に身長を描画した折れ線グラフを描画してください。その際に、グラフはクラスごとのグラフにして、それぞれに組名のラベルをつけて、凡例を付与してください。」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」・「各種グラフのオプション設定方法」節
</details>

```python
fig = plt.figure()
ax = fig.add_subplot(111)

for class_type in data['クラス'].unique():
  ax.plot(data.loc[data['クラス'] == class_type, '身長'].index, data.loc[data['クラス'] == class_type, '身長'].values, label=class_type)

ax.legend()
plt.show()
```

### 1.10 「dataの『身長』列の分布を確認するために、クラスA・クラスBの『身長』列についてそれぞれのヒストグラムを描画してください。その際に、グラフの透明度を半透明にしてください。(透明度は好きな値に設定してください。)」

<details>
<summary>
この問題に関連する内容を復習するためのスライド
</summary>

Matplotlibの「グラフ作成手順」・「各種グラフの作成方法」・「各種グラフのオプション設定方法」節
</details>

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(data.loc[data['クラス'] == 'a', '身長'].values, label='a', alpha=0.5)
ax.hist(data.loc[data['クラス'] == 'b', '身長'].values, label='b', alpha=0.5)

ax.legend()
plt.show()
```

### 1.13 「簡潔な描画方法を用いて、1.9と同様のオプションを付与した折れ線グラフを描画してください。」

### 1.14 「seabornを用いて、dataの『身長』列のヴァイオリンプロットを描画してください。」
